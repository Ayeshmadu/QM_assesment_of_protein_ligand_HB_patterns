```python
#imports
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
```

#### Rename Columns

```python
df.columns = ["D_sy", "H_Id", "Acc_id", "A_sy", "D-H...A", "length", "angle", "A_F", 
              "D_F", "A_fr", "D_fr", "Intra", "BIFR", "D_re", "A_re", "Lig1" ]
```

```python
df.head(22)
```

#### Remove MOL and MOL (Intramolecular) and few changes

```python
df2 = df[df['A_re'] != df['D_re']]
```

```python
df2["Lig1"]=df2["Lig1"].replace('OC','CO')
```

```python
df2["Lig1"]=df2["Lig1"].replace('OS','SO')
```

```python
df2["Lig1"]=df2["Lig1"].replace('ON','NO')
```

```python
df2["Lig1"]=df2["Lig1"].replace('NC','CN')
```

```python
df2["Lig1"]=df2["Lig1"].replace('OP','PO')
```

```python
#df2 = df2.drop(labels=20, axis=0)
df2 = df2[(df[['A_F']] > 0).all(1)]
```

```python
df2 = df2[(df[['A_F']] < 0.6).all(1)]
```

```python
df2.to_csv('QSAR_new.csv')
```
